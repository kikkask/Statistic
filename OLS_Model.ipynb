**OLS Model of the relationship between the area of ​​houses (m²) and the rental value in a given city**

by: Lucas Brito Cordeiro

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
```

```python
df = pd.read_csv('/content/dataset.csv')
df.head()
```

```python
# Definition of the dependent variable
y = df['valor_aluguel']

# Definition of the independent variable
x = df['area_m2']

x = sm.add_constant(x)

# Creating the model
model = sm.OLS(y, x)
```

```python
result = model.fit()
```

```python
print(result.summary())
```

```python
plt.figure(figsize=(10, 6))
plt.xlabel('area_m2', size = 16)
plt.ylabel('valor_aluguel', size = 16)
plt.plot(x['area_m2'], y, 'o', label = 'Real Data')
plt.plot(x['area_m2'], result.fittedvalues, 'r-', label = 'Regresion Line')
plt.legend(loc = 'best')
plt.show()
```

**Conclusion:**

P> |T| = 0.000
R² = 0.342

So, yes. Area helps explain rental price. However, it is not the only variable to explain the rental price, considering that 0.342 of R² is still far from 1.

The ideal would be to create a multiple linear regression model, using more variables to find a stronger relationship with the aim of the machine understands how the rental price can be generated and making more accurate price predictions.
